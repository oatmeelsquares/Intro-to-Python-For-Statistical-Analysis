### 4.1 Data Types

There are three core data types in python:

- Numeric (`int`, `float`)
    - `int` represents an integer: any whole number, including negatives and zero
    - `float` represents any number with a decimal
- String (`str`)
    - a series of characters bounded by quotation marks (single or double)
    - adding strings concatenates one after the other (strings of numbers will not be evaluated as numeric)
- Boolean (`bool`)
    - a truth value, either `True` or `False`
    - capitalization is important!
    - `T` and `F` don't work
    - Boolean operators:
        - `==` to check if two things are equal
        - `>` or `>=` to check if one thing is less than (or equal to) another thing
        - `<` or `<=` to check if one thing is greater than (or equal to) another thing
        - `!=` to check if two things are _unequal_
        - `in` (for lists or sets) to  check if something can be found in a given list or set
    - boolean expressions can be combined with `and` and `or`
    - `True` can be interpreted as $1$ and `False` can be interpreted as `0`
    - If converting the other way around, $0$ will map to `False` and any nonzero number will map to `True`

#### 4.2.1 Methods and Attributes

Methods and attributes are inherent to a certain kind of data or structure.
- An attribute is some *thing* inherent to a certain kind of data or structure.
    - *e.g.* all dictionaries have keys and values
    - attributes are objects like strings, numbers, or lists
    - can be any data type or structure
- A method is something that a certain kind of data or structure can inherently *do*.
    - *e.g.* every list can be added to
    - a method is a function, not an object
    - must be accessed (called) using parenthesis

Methods and attributes of an object can be accessed using dot notation
- **This is different from in R**
- If you type a dot `.` after a word, it always means you're accessing something inside of the variable represented by that word
- You can't put dots in names like you can in R

````python
my.name = 'Becky' # doesn't work - 'name' is assumed to already be part of something called 'my'
````

## 5 `if`, Loops and indentation

- `if` statements are used to create branches in the flow of your code
    - the code in the `if` statement block will only execute if the condition is met
    - can be used with or without `elif` and `else`
- `for` loops are for iterating through lists, etc. of known length
    - it is common to use a `range` with `for` loops
    - `range(a, b, c)` gives you an object like a list (but not a list) of every number between `a` and `b` (not including `b`) with a step size of `c`
- `while` loops are for continuing an action an unknown number of times based on some stop condition

**Indentation**
- Indentation is good practice in all languages, but is necessary in python
- If you're coding a loop, eveything in that loop must be indented
- (this is in lieu of needing brackets around the loop - the interpreter knows when to end the loop based on when the indentation stops)
- You can indent as much or as little as you want, but you have to be consistent
- You can use `Tab` to help stay consistent

#### `.loc`

- `.loc` is a method of the pandas DataFrame type of object
- Works like indexing
- Works with parenthesis or square brackets
- Can take lists of indices (whether column names or index strings)
- Can take slices if indices are numeric, which are inclusive of both bounds
- If you want to refer to the actual numeric positions, use `.iloc`

````python
columns_I_want = ['ReceivedDBTIntervention', 
                  'TeacherSameEthnicity',
                  'FreeReducedSchoolLunchGroup',
                  'TotalNumberofAdultsstHome',
                  'Female',
                  'MathGrade'] # a list of column names
df.loc[5:10, columns_I_want]
````

````python
# notice the subtle difference of iloc
df.iloc[5:10]
````

N.B. The above dataframe actually starts at row position 5 (because of zero-indexing), which is ID 6. The last row is position 9 (slice 5:10 does not include position 10), which is ID 10 because the ID starts at 1.

````python
data = df.loc[:, ['TotalNumberofAdultsstHome', 'MathGrade']] # notice square brackets for list of column names

data.columns = ['adults', 'grade'] # rename all columns manually
data
````
